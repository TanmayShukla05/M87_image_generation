```python
from pyuvdata import UVData

uv = UVData()
uv.read_uvfits('M87_combined_all.uvfits')

# Print basic summary
print(uv)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(uv.uvw_array[:,0], uv.uvw_array[:,1], s=1, alpha=0.5)
plt.xlabel('u (m)')
plt.ylabel('v (m)')
plt.title('UV Coverage')
plt.grid(True)
plt.axis('equal')
plt.show()
```
